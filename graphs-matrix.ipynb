```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


# Example classification report values
report = {
    0: {'Recall': 0.98, 'Precision': 0.97, 'F1-Score': 0.98, 'Support': 9648},
    1: {'Recall': 0.97, 'Precision': 0.98, 'F1-Score': 0.97, 'Support': 8917}
}
# Define the starting and ending colors
start_color = '#FFFFFF'
end_color = '#00b300'

# Generate the color map
cmap_colors = []
cmap_colors = ['#FFFFFF']*9000
for i in range(1000):
    r = int((i/999) * int(end_color[1:3], 16) + ((999-i)/999) * int(start_color[1:3], 16))
    g = int((i/999) * int(end_color[3:5], 16) + ((999-i)/999) * int(start_color[3:5], 16))
    b = int((i/999) * int(end_color[5:], 16) + ((999-i)/999) * int(start_color[5:], 16))
    cmap_colors.append(f'#{r:02X}{g:02X}{b:02X}')

cmap = ListedColormap(cmap_colors[:len(cmap_colors)-1])
columns = ['Recall', 'Precision', 'F1-Score', 'Support']
classes = ['Normal (0)', 'Attack (1)']

# Create a table with the values
cell_text = []
for row in range(len(classes)):
    cell_text.append([report[row][column] for column in columns])

# Create the figure and axis
fig, ax = plt.subplots(figsize=(7, 3))
ax.axis('off')

# Add the topic label
fig.text(0.5, 0.95, 'CNN Classification Report',
            fontsize=16, ha='center')

# Create the table and format it
table = ax.table(cellText=cell_text,
                    colLabels=columns,
                    rowLabels=classes,
                    cellLoc='center',
                    bbox=[0, 0, 1, 1],
                    colWidths=[0.2, 0.2, 0.2, 0.2],
                    cellColours=[[cmap_colors[0]] * len(columns)] + [[cmap_colors[1]] * len(columns)])

table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)

# Adjust cell padding and font properties
for key, cell in table.get_celld().items():
    cell.set_text_props()
    if key[0] == 0:
        cell.set_fontsize(12)
    else:
        cell.set_fontsize(16)
    cell.set_height(0.15)
    cell.set_edgecolor('white')

# Color the cells using the colormap (for the first 3 columns)
for i, key in enumerate(columns[:3]):
    cell_values = np.array([report[row][key]
                            for row in range(len(classes))])
    colors = cmap(cell_values / cell_values.max())
    for j in range(len(classes)):
        table.get_celld()[j+1, i].set_facecolor(colors[j])

# Set the last column background color to white
for j in range(len(classes)):
    table.get_celld()[j+1, len(columns)-1].set_facecolor('white')

# Add a colorbar with decimal number labels
colorbar = fig.colorbar(plt.cm.ScalarMappable(
    cmap=cmap), ax=ax, boundaries=np.linspace(0.9, 1, 1000))
colorbar.set_ticks(np.linspace(0.9, 1, 10))
colorbar.set_ticklabels(
    [f'{i:.2f}' for i in np.linspace(0.9, 1, 10)])
colorbar.ax.yaxis.set_ticks_position('right')

# Display the figure
plt.tight_layout()
plt.show()
```
