```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Define the confusion matrix
conf_matrix = np.array([[9485, 163],
                        [285, 8632]])

# Define the color map
cmap = ListedColormap(['white', '#00b300', '#00b300'])

# Plot the confusion matrix as a colored table
fig, ax = plt.subplots()
im = ax.imshow(conf_matrix, cmap=cmap)

# Set the axis labels
ax.set_xticks(np.arange(2))
ax.set_yticks(np.arange(2))
ax.set_xticklabels(['Normal (0)', 'Attack (1)'], fontsize=15)
ax.set_yticklabels(['Normal (0)', 'Attack (1)'], fontsize=15)
ax.tick_params(axis='both', which='both', length=0)

# Add the values inside the cells
for i in range(2):
    for j in range(2):
        text = ax.text(j, i, conf_matrix[i, j], ha="center", va="center", color="black", fontsize=30)

# Set the plot title
ax.set_title('CNN Confusion Matrix', fontsize=18)

# Add axis labels
plt.xlabel('Predicted', fontsize=15)
plt.ylabel('Actual', fontsize=15)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
ind = np.arange(N)
width = 0.105

CNN  = [97.59,97.79,97.19,97.96,97.49]
MLP  = [97.64,97.71,97.39,97.88,97.55]
RNN  = [96.77,98.20,95.04,98.39,96.59]
LSTM = [96.74,97.97,95.20,98.17,96.56]
GRU  = [96.67,98.20,94.83,98.38,96.48]
GAN  = [96.98,97.51,96.20,97.71,96.85]
SVM  = [96.89,97.61,95.90,97.81,97.06]
XGB  = [98.03,98.48,97.43,98.59,97.95]

bar1 = plt.bar(ind, CNN, width, color = 'purple')
bar2 = plt.bar(ind+width, MLP, width, color='blue')
bar3 = plt.bar(ind+width*2, RNN, width, color = 'skyblue')
bar4 = plt.bar(ind+width*3, LSTM, width, color = 'green')
bar5 = plt.bar(ind+width*4, GRU, width, color = 'yellow')
bar6 = plt.bar(ind+width*5, GAN, width, color = 'orange')
bar7 = plt.bar(ind+width*6, SVM, width, color = 'red')
bar8 = plt.bar(ind+width*7, XGB, width, color = 'violet')

plt.xlabel("Evaluation Metrics")
plt.ylabel('Score %')
plt.title("Model Accuracy Metrics Comparision")

plt.xticks(ind+width*3.5,['Accuracy Score', 'Precision', 'Recall', 'Specificity', 'F1 Score'])
plt.legend((bar1, bar2, bar3, bar4, bar5, bar6, bar7, bar8), ('CNN', 'MLP', 'RNN', 'LSTM', 'GRU', 'GAN', 'SVM', 'XGBoost'), loc="upper right" )
plt.ylim(93,100.6)
plt.gcf().set_size_inches(14, 8)
#plt.axhline(y = 100, color = 'grey', linestyle = '--')

plt.show()
```
